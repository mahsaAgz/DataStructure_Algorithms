## Hash Table (Dictionary in Python)

### Definition:

A hash table is a data structure that implements an associative array, a structure that can map keys to values. In Python, the dictionary (`dict`) is a built-in implementation of a hash table.

### Characteristics:

- **Key-Value Pairs:** Stores data in pairs where each key is unique.
- **Hash Function:** Uses a hash function to compute an index into an array of buckets or slots, from which the desired value can be found.
- **Unordered:** The elements are not stored in a particular order.

### Operations and Time Complexity:

- **Insert/Update (`dict[key] = value`):** O(1) average, O(n) worst case
- **Delete (`del dict[key]`):** O(1) average, O(n) worst case
- **Lookup (`dict[key]`):** O(1) average, O(n) worst case

### Use Cases:

- Fast lookups, inserts, and deletes based on key.
- Implementing associative arrays, caches, and sets.

```python
my_dict = {}
print(my_dict)

my_dict['apple'] = 1  # Insert
my_dict['banana'] = 2
print(my_dict)

value = my_dict['apple']  # Lookup
print(value)

del my_dict['banana']  # Delete
print(my_dict)
```

## Heap (Priority Queue)

### Definition:

A heap is a specialized tree-based data structure that satisfies the heap property. In a max heap, for any given node, the value of the node is greater than or equal to the values of its children. In a min heap, the value of the node is less than or equal to the values of its children. Heaps are commonly used to implement priority queues.

### Characteristics:

- **Complete Binary Tree:** Heaps are usually implemented as a binary heap, which is a complete binary tree.
- **Heap Property:** In a min heap, each parent node is less than or equal to its children. In a max heap, each parent node is greater than or equal to its children.

### Operations and Time Complexity:

- **Insert:** O(log n)
- **Delete Min/Max:** O(log n)
- **Peek Min/Max:** O(1)

### Use Cases:

- Implementing priority queues.
- Efficiently finding the minimum or maximum element.
- Heap sort algorithm.

### Implementation in Python:

Python's `heapq` module provides an implementation of the heap queue algorithm (priority queue).

```python
import heapq

min_heap = []
print(min_heap)
heapq.heappush(min_heap, 3)  # Insert
heapq.heappush(min_heap, 1)
heapq.heappush(min_heap, 2)
print(min_heap)
smallest = heapq.heappop(min_heap)  # Delete Min (1)
print(min_heap)
print(smallest)  # Output: 1
```

## Conclusion

- **Hash Table (Dictionary):** Efficient for fast lookups, inserts, and deletes based on keys with average O(1) time complexity. Suitable for use cases where quick access to elements by key is required.
- **Heap (Priority Queue):** Efficient for operations where the minimum or maximum element needs to be quickly accessed or removed, with O(log n) time complexity for insertions and deletions. Ideal for implementing priority queues and algorithms like heap sort.
